```python
marks = []

for i in range(10):
    marks.append(int(input("Enter the marks for 10 students:")))
average= sum(marks)/10
highest= max(marks)
lowest = min(marks)
count=0
print("Highest marks :", highest)
print("Lowest marks :", lowest)
print("Average marks :", average)
for score in marks:
    if score>average: count+=1
print("No. of students considered as above average :", count)
gradedist={"A":0,"B":0,"C":0,"D":0}
for score in marks:
    if score>=90: gradedist["A"]+=1
    elif score>=75 and score<90: gradedist["B"]+=1
    elif score>=60 and score<75: gradedist["C"]+=1
    else : gradedist["D"]+=1
print("GRADE DISTRIBUTION :", gradedist)



```
