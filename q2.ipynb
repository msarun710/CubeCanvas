```python
face = [
    ["R","R","G"],
    ["W","R","G"],
    ["B","B","Y"]
]
 
for row in face:
    for value in row:
        print(value, end=" ")
    print()      
counts={}
for row in face:
    for value in row:
        if value in counts:
            counts[value]+=1
        else: counts[value]=1
print(counts)
for row in face:
    for index in range(len(row)):
        if row[index]=="R": row[index]="X" 
print(face)        

highestcount=0        
mostcomcolor=""
for color in counts:
    if counts[color]>highestcount:
        highestcount= counts[color]
        mostcomcolor=color
print("The most common color is:", mostcomcolor)
nrows=0
for row in face:
    if row[0]==row[1] and row[0]==row[2]:
        nrows+=1
if nrows>0:print("YES")
else: print("NO")



```
