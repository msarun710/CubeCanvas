```python
sentence=input("Enter a sentence:")
nofchar= len(sentence)
print("Total characters:", nofchar)
nofwords=0;
for char in sentence:
    if char==" ": nofwords+=1
if nofwords>0: 
    print("No. of words :", nofwords+1)    

vowels="aeiouAEIOU"
nofvowels=0
for char in sentence:
    if char in vowels:
        nofvowels+=1
    
print("No. of vowels :",nofvowels)    
dict={}
for char in sentence:
    if char in dict: dict[char]+=1
    else: dict[char]=1
print(dict)
mostfreqc=""
highestcount=0
for char in dict:
    if dict[char]>highestcount: 
        highestcount=dict[char]
        mostfreqc=char
print("The most frequent character is :",mostfreqc)
modstr=""
for char in sentence:
    if char!=" ":
        modstr=modstr+char.lower()
revstr=""
for char in sentence:
    if char!=" ":
        revstr=char.lower()+ revstr
if revstr==modstr: print("Palindrome : TRUE")
else: print("Palindrome : FALSE")        
        
```
