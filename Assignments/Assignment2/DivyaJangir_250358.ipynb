# Question 4: Interpolation Investigation (Hard)

Different interpolation techniques produce different results.


## Part A (Theory)

Research and explain:

1. Nearest Neighbor Interpolation
2. Bilinear Interpolation
3. Bicubic Interpolation
4. Lanczos Interpolation

For each method discuss:
- Working principle
- Advantages
- Disadvantages
- Suitable applications


### Theory Answer Space

Write your answer here.

1. Nearest Neighbor Interpolation assigns the new pixel location the value of closest pixel location. Its advantages are that it is fast, preserves sharp edges and is good for pixel art. Disadvantages are that it lowers the image quality as compared to other methods. One of the suitable applications is that it can be used in pixel art.
2. Bilinear Interpolation uses four neighbouring pixels to calculate the value of new pixel. Its advantage is that it produces smoother results. Disadvantage are that it does not preserve sharp details perfectly and also can produce slight blurring. Suitable applications are that it can be used for video processing and general image resizing.
3. Bicubic Interpolation uses 16 neighbouring pixels to produce to  assign the value of new pixel. Advantages are that it gives better quality, smooth transitions and reduce artifacts. Disadvantages are that it is slower than nearest neighbour and bilinear methods, and also requires more computational resources. It can be used for professional photo editing and high quality image scaling.
4. Lanczos Interpolation uses sinc function to calculate the new pixel value. It produces image of excellent quality, sharp details and minimal aliasing. Disadvantages are that it is computationally expensive and may create ringing artifacts near sharp edges. It is used for high-quality image enlargment, digital photography, professional image and video processing.
