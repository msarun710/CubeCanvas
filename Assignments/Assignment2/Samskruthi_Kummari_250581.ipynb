## Part B (Coding)

   - 300 × 300

5. Save all generated images.


````python
from PIL import Image

img = Image.open("Donald Trump (1).jpeg")
````

````python
width, height = img.size
print("Original Dimensions:", width, "x", height)
````

````python
left = width // 4
top = height // 4
right = left + width // 2
bottom = top + height // 2

cropped = img.crop((left, top, right, bottom))
````

````python
img_300 = cropped.resize((300, 300))
img_600 = cropped.resize((600, 600))
````

````python
cropped.save("cropped_image.jpeg")
img_300.save("resized_300x300.jpeg")
img_600.save("resized_600x600.jpeg")
````

# Question 4: Interpolation Investigation (Hard)

Different interpolation techniques produce different results.


## Part A (Theory)

Research and explain:

1. Nearest Neighbor Interpolation
2. Bilinear Interpolation
3. Bicubic Interpolation
4. Lanczos Interpolation

For each method discuss:
- Working principle
- Advantages
- Disadvantages
- Suitable applications


### Theory Answer Space

Write your answer here.

1. Nearest Neighbor Interpolation

  Working Principle:

  Nearest Neighbor chooses the value of the closest pixel and copies it to the
  
  new pixel when resizing an image.

  Advantages:
  
  Very fast.
  
  Easy to use.
  
  Requires less computer processing.
  
  Disadvantages:
  
  Image can look blocky or pixelated.
  
  Lower image quality when enlarged.

  Suitable Applications:
  
  Pixel art.
  
  Simple graphics.
  
  Applications where speed is more important than quality.




2. Bilinear Interpolation
   
   Working Principle:

   Bilinear Interpolation looks at the 4 nearest pixels and calculates an
   
   average value to create a new pixel.

   Advantages:

   Produces smoother images than Nearest Neighbor.

   Reduces pixelation.
   
   Good balance between speed and quality.
   
   Disadvantages:
   
   Image may look slightly blurry.
   
   Fine details can be lost.
   
   Suitable Applications:
   
   General image resizing.
   
   Websites and digital images.
   
   Everyday image processing


3. Bicubic Interpolation
   
   Working Principle:

   Bicubic Interpolation uses the nearest 16 surrounding pixels to calculate
   
   the value of a new pixel. This creates smoother and sharper results.

   Advantages:
   
   Better image quality.
   
   Preserves more details.
   
   Produces smoother images.
   
   Disadvantages:
   
   Slower than Nearest Neighbor and Bilinear.
   
   Uses more computer resources.
   
   Suitable Applications:
   
   Photo editing.
   
   Printing images.
   
   High-quality image resizing.



4. Lanczos Interpolation
   
   Working Principle:

   Lanczos Interpolation uses many nearby pixels and a complex mathematical
   
   formula to create the new pixel values. It aims to keep as much detail as
   
   possible.

   Advantages:
   
   Very high image quality.
   
   Preserves fine details.
   
   Produces sharp images.
   
   Disadvantages:
   
   Slowest method.
   
   Requires more processing power.
   
   May create small edge artifacts in some images.
   
   Suitable Applications:
   
   Professional photography.
   
   Graphic design.
   
   High-resolution image enlargement.

## Part B (Coding)

Using the same image:

1. Resize the image to 100x100.
2. Resize it back to 800x800 using:
   - Nearest Neighbor
   - Bilinear
   - Bicubic
   - Lanczos
3. Save all outputs.


````python
# Load image
img = Image.open("Donald Trump (1).jpeg")

````

````python
# Step 1: Resize to 100x100
small_img = img.resize((100, 100))
small_img.save("small_100x100.jpeg")
````

````python
# Step 2: Resize back to 800x800 using different methods

# Nearest Neighbor
nearest_img = small_img.resize((800, 800), Image.Resampling.NEAREST)
nearest_img.save("nearest_800x800.jpeg")

# Bilinear
bilinear_img = small_img.resize((800, 800), Image.Resampling.BILINEAR)
bilinear_img.save("bilinear_800x800.jpeg")

# Bicubic
bicubic_img = small_img.resize((800, 800), Image.Resampling.BICUBIC)
bicubic_img.save("bicubic_800x800.jpeg")

# Lanczos
lanczos_img = small_img.resize((800, 800), Image.Resampling.LANCZOS)
lanczos_img.save("lanczos_800x800.jpeg")

print("All images saved successfully!")
````
