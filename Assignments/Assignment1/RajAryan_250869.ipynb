## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
maxm, minm, sum = 0, 100, 0
marks = []
for x in range(10):
    n = int(input(f"Enter the marks of student {x+1}: "))
    while n<0 or n>100:
        n = int(input("Marks should be from 0 to 100: "))
    marks.append(n)
    maxm = max(maxm, n)
    minm = min(minm, n)
    sum += n

avg = sum/10
aboveAvg = 0
grades = {
    "A": 0,
    "B": 0,
    "C": 0,
    "D": 0
}
for i in marks:
    if i > avg:
        aboveAvg += 1
    
    if i>=90:
        grades["A"]+=1
    elif i>=75:
        grades["B"]+=1
    elif i>=60:
        grades["C"]+=1
    else:
        grades["D"]+=1

print("Max Marks: ", maxm)
print("Min Marks: ", minm)
print("Avg Marks: ", avg)
print("Students with above avg Marks: ", aboveAvg)
for a, b in grades.items():
    print(a, ": ", b)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]

validColors = {
    "R":0,
    "O":0,
    "G":0,
    "B":0,
    "W":0,
    "Y":0
}

solvedRow = []

i=0
for row in face:
    for color in row:
        if color not in validColors:
            print("(Invalid)", end=" ")
        else:
            print("X" if color=="R" else color, end=" ")
            validColors[color]+=1
    print("")
    i+=1
    if(len(set(row))==1):
        solvedRow.append(i)

highestFreq = max(validColors, key=validColors.get)

if len(solvedRow)==0:
    print("No row has all identical colors.")
else:
    for x in solvedRow:
        print("Row ", x, " has all identical colors.")

print("Most frequent color is: ", highestFreq)
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
vowels = ('a', 'e', 'i', 'o', 'u')
chars = {}
sentence = input("Enter the sentence: ")
wordsList = sentence.split()
numWords, numChar, numVowels = len(wordsList), 0, 0

for word in wordsList:
    numChar += len(word)
    for ch in word:
        if ch.lower() in vowels:
            numVowels+=1

        if ch.lower() in chars:
            chars[ch.lower()]+=1
        else:
            chars[ch.lower()]=1

noSpaces = "".join(wordsList)
noSpacesRev = noSpaces[::-1]
mostFreq = max(chars, key=chars.get)


if noSpaces.lower() == noSpacesRev.lower():
    print("Sentence is a palindrome.")
else:
    print("Sentence is not a palindrome.")

print("Number of words is: ", numWords)
print("Number of characters is: ", numChar)
print("Number of vowels is: ", numVowels)
print("Most frequent character is: ", mostFreq)
````

# Mini Project (50 Marks): Speedcubing Club Management System

The Speedcubing Club wants a software system that can be reused by future coordinators to manage members, attendance and events.

## Class: Member

Each member should store:
- Name
- Roll Number
- Branch
- Year
- Cubing Average

Methods:
- `display_details()`
- `update_average()`

## Class: Event

Each event should store:
- Event Name
- Date
- Venue
- List of Registered Members

Methods:
- `register_member()`
- `remove_member()`
- `display_participants()`

## Class: Club

The Club class should:

- Store all members.
- Store all events.
- Add a member.
- Remove a member.
- Search for a member using roll number.
- Create an event.
- Display all events.
- Display all members.
- Display the top 3 fastest cubers in the club.
- Find average cubing time of all members.
- Count members branch-wise.

## Bonus Features

Implement any two:

1. Attendance tracking for workshops.
2. Member leaderboard based on cubing averages.
3. Event-wise participant statistics.
4. Search members by branch.
5. Export member information in a readable text format.

## Reflection Questions

Answer the following:

1. Where did you use encapsulation?
2. Which classes interact with each other and why?
3. Which variables are instance variables?
4. Which variables could be class variables?
5. What modifications would be required to reuse your system for next year?


````python
# Mini Project Solution
class Member:
    def __init__(self, name, rollno, branch, year, cubeAvg):
        self.name = name
        self.rollno = rollno
        self.branch = branch
        self.year = year
        self.__cubeAvg = cubeAvg
    
    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.rollno}")
        print(f"Branch: {self.branch}")
        print(f"Year: {self.year}")
        print(f"Cubing Avg: {self.__cubeAvg}")

    def update_average(self, newAvg):
        self.__cubeAvg = newAvg
        print(f"Cubing average for {self.name} updated to {self.__cubeAvg}")

    def get_average(self):
        return self.__cubeAvg


class Event:
    def __init__(self, eventName, date, venue):
        self.eventName = eventName
        self.date = date
        self.venue = venue
        self.regMembers = []

    def register_member(self, memberObj):
        if memberObj in self.regMembers:
            print(f"{memberObj.name} is already registered.")
        else:
            self.regMembers.append(memberObj)
            print(f"{memberObj.name} has been registered for {self.eventName}.")

    def remove_member(self, memberObj):
        if memberObj in self.regMembers:
            self.regMembers.remove(memberObj)
            print(f"{memberObj.name} has been removed from {self.eventName}.")
        else:
            print(f"{memberObj.name} hasn't been registered.")

    def display_participants(self):
        if not self.regMembers:
            print("No registered members yet.")
        else:
            for member in self.regMembers:
                print(member.name, ": ", member.rollno)

class Club:
    def __init__(self):
        self.members = {}
        self.events = {}

    def addMember(self, name, rollno, branch, year, cubeAvg):
        self.members[rollno] = Member(name, rollno, branch, year, cubeAvg)
        print(f"{name} (Roll: {rollno}) has been added.")

    def removeMember(self, rollno):
        if rollno in self.members:
            del self.members[rollno]
            print(f"Member with roll number {rollno} has been removed.")
        else:
            print(f"No member with roll number {rollno}.")

    def searchMember(self, rollno):
        if rollno in self.members:
            self.members[rollno].display_details()
        else:
            print(f"No member with roll number {rollno}.")

    def displayAll(self):
        if not self.members:
            print("No members in the club.")
        else:
            for rollno, member in self.members.items():
                print(f"{member.name}: {rollno}")        

    def overallAvg(self):
        if not self.members:
            print("No members in the club.")
        else:
            sumOfAvg=0
            for member in self.members.values():
                sumOfAvg += member.get_average()
            avgOfAll = sumOfAvg/len(self.members)
            print(f"Overall avg is {avgOfAll:.2f} seconds.")

    def branchCount(self):
        if not self.members:
            print("No members in the club.")
        else:
            branchList = {}
            for member in self.members.values():
                if member.branch in branchList:
                    branchList[member.branch]+=1
                else:
                    branchList[member.branch]=1
            
            for branch, count in branchList.items():
                print(f"{branch}: {count} members")

    def top3Avg(self):
        if not self.members:
            print("No members in the club.")
        else:
            cubeAvgList = []
            for m in self.members.values():
                # Updated to use the safe getter method
                cubeAvgList.append((m.get_average(), m.name))
            
            cubeAvgList.sort()
            if len(cubeAvgList)<3:
                for i in range(len(cubeAvgList)):
                    print(f"{i+1}. {cubeAvgList[i][0]}, {cubeAvgList[i][1]}")
            else:
                for i in range(3):
                    print(f"{i+1}. {cubeAvgList[i][0]}, {cubeAvgList[i][1]}")

    def createEvent(self, event_name, date, venue):
        newEvent = Event(event_name, date, venue)
        self.events[event_name] = newEvent
        print(f"Event '{event_name}' created.")

    def displayEvents(self):
        if not self.events:
            print("No events created yet.")
        else:
            for event_name, event_obj in self.events.items():
                print(f"{event_name} | Date: {event_obj.date} | Venue: {event_obj.venue}")

    def searchByBranch(self, branch):
        print(f"Members in {branch}:")
        found = False
        for member in self.members.values():
            if member.branch.upper() == branch.upper():
                member.display_details()
                found = True
        if not found:
            print(f"No members found in {branch}.")

    def leaderboard(self):
        if not self.members:
            print("No members in the club.")
        else:
            cubeAvgList = []
            for m in self.members.values():
                # Updated to use the safe getter method
                cubeAvgList.append((m.get_average(), m.name))
            
            cubeAvgList.sort()
            for i in range(len(cubeAvgList)):
                print(f"{i+1}. {cubeAvgList[i][0]}, {cubeAvgList[i][1]}")
````
