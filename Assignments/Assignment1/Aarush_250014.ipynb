## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
def analyze_marks():
    marks = []
    num_students = 10
    
    # Taking input 
    print(f"Please enter the marks of {num_students} students (0-100):")
    
    # Storing the input in a list
    student_count = 0
    while student_count < num_students:
        try:
            mark = float(input(f"Student {student_count + 1}: "))
            if mark >= 0 and mark <= 100:
                marks.append(mark)
                student_count += 1
            else:
                print("Marks should be between 0 and 100. Try again.")
        except ValueError:
            print("Invalid input. Please enter a numerical value.")

    # Calculating highest, lowest, average and above average marks
    highest_marks = marks[0]
    lowest_marks = marks[0]
    total_sum = 0.0
    
    for mark in marks:
        if mark > highest_marks:
            highest_marks = mark
            
        if mark < lowest_marks:
            lowest_marks = mark
            
        total_sum = total_sum + mark
        
    average_marks = total_sum / num_students
    
    above_avg_count = 0
    for mark in marks:
        if mark > average_marks:
            above_avg_count = above_avg_count + 1

    # Assigning grades
    count_A = 0
    count_B = 0
    count_C = 0
    count_D = 0
    
    for mark in marks:
        if mark >= 90:
            count_A = count_A + 1
        elif mark >= 75:
            count_B = count_B + 1
        elif mark >= 60:
            count_C = count_C + 1
        else:
            count_D = count_D + 1

    # Output
    grade_distribution = {
        'A': count_A,
        'B': count_B,
        'C': count_C,
        'D': count_D
    }

    print(f"Highest Marks  : {highest_marks}")
    print(f"Lowest Marks   : {lowest_marks}")
    print(f"Average Marks  : {average_marks:.2f}")
    print(f"Above Average  : {above_avg_count} student(s)")
    
    print("\nGrade Distribution ")
    print(f"Grade A (Count) : {grade_distribution['A']}")
    print(f"Grade B (Count) : {grade_distribution['B']}")
    print(f"Grade C (Count) : {grade_distribution['C']}")
    print(f"Grade D (Count) : {grade_distribution['D']}")

if __name__ == "__main__":
    analyze_marks()
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
def process_grid():
    # Making a 2D List
    face = [
        ["R", "R", "G"],
        ["W", "R", "G"],
        ["B", "B", "Y"]
    ]

    rows = 3
    cols = 3

    print(" Original Grid ")

    for r in range(rows):
        for c in range(cols):
            print(face[r][c], end=" ")
        print()

    # Counting frequencies
    frequencies = {}
    for r in range(rows):
        for c in range(cols):
            current_color = face[r][c]
            # New colour
            if current_color not in frequencies:
                frequencies[current_color] = 1
            else:
                # Existing colour
                frequencies[current_color] = frequencies[current_color] + 1

    print("\n Colour Frequencies ")
    for color in frequencies:
        print(f"{color}: {frequencies[color]}")

    # Most common colour
    most_common_color = ""
    highest_count = 0
    
    for color in frequencies:
        if frequencies[color] > highest_count:
            highest_count = frequencies[color]
            most_common_color = color
            
    print(f"\nMost Common Colour: {most_common_color} (Appears {highest_count} times)")

    # Replacing R with X
    for r in range(rows):
        for c in range(cols):
            if face[r][c] == "R":
                face[r][c] = "X"

    print("\nFinal Grid R -> X:")
    for r in range(rows):
        for c in range(cols):
            print(face[r][c], end=" ")
        print()

    # Checking if any row has all identical colours
    identical_row_found = False
    
    for r in range(rows):
        first_color = face[r][0]
        row_is_identical = True
        
        # Next column
        for c in range(1, cols):
            if face[r][c] != first_color:
                row_is_identical = False
                break 
                
        if row_is_identical == True:
            identical_row_found = True
            break

    print(f"\nAny row has all identical colors? {identical_row_found}")

if __name__ == "__main__":
    process_grid()
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
def text_analytics():
    text = input("Enter a sentence: ")

    # Total Characters
    total_chars = len(text)

    # Total words
    total_words = 0
    in_word = False
    
    for char in text:
        # If the character is not a space, tab, or newline
        if char != ' ' and char != '\t' and char != '\n':
            if in_word == False:
                total_words += 1
                in_word = True  # New word found
        else:
            in_word = False # Word ended

    # Counting vowels
    vowels_count = 0
    for char in text:
        if (char == 'a' or char == 'A' or 
            char == 'e' or char == 'E' or 
            char == 'i' or char == 'I' or 
            char == 'o' or char == 'O' or 
            char == 'u' or char == 'U'):
            vowels_count += 1

    # Counting frequency
    char_freq = {}
    for char in text:
        if char not in char_freq:
            char_freq[char] = 1
        else:
            char_freq[char] += 1

    # Highest frequency
    max_char = ''
    max_freq = 0
    for char in char_freq:
        if char_freq[char] > max_freq:
            max_freq = char_freq[char]
            max_char = char

    # Palindrome check
    clean_chars = []
    clean_len = 0
    for char in text:
        if char != ' ':
            ascii_val = ord(char)
            # Case handling
            if ascii_val >= 65 and ascii_val <= 90:
                clean_chars.append(chr(ascii_val + 32))
            else:
                clean_chars.append(char)
            clean_len += 1

    is_palindrome = True
    left = 0
    right = clean_len - 1
    
    while left < right:
        if clean_chars[left] != clean_chars[right]:
            is_palindrome = False
            break # Exit loop early if mismatch found
        left += 1
        right -= 1

    # Output
    print(f"1. Total Characters: {total_chars}")
    print(f"2. Total Words:      {total_words}")
    print(f"3. Total Vowels:     {vowels_count}")
    
    print("4. Character Frequencies:")
    for char in char_freq:
        print(f"   '{char}' : {char_freq[char]}")
        
    print(f"5. Most Frequent: '{max_char}' ({max_freq} times)")
    print(f"6. Palindrome: {is_palindrome}")

if __name__ == "__main__":
    text_analytics()
````

# Mini Project (50 Marks): Speedcubing Club Management System

The Speedcubing Club wants a software system that can be reused by future coordinators to manage members, attendance and events.

## Class: Member

Each member should store:
- Name

## Class: Event

## Class: Club

The Club class should:

- Store all members.
- Store all events.
- Add a member.
- Remove a member.
- Search for a member using roll number.
- Create an event.
- Display all events.
- Display all members.

- Find average cubing time of all members.
- Count members branch-wise.

## Bonus Features

Implement any two:

````python
# Mini Project Solution
class MemberStruct:
    def __init__(self, n_str, r_str, b_str, y_int, c_flt):
        self.name_str = n_str
        self.roll_no_str = r_str
        self.branch_str = b_str
        self.year_int = y_int
        self.avg_time_flt = c_flt

    def print_details(self):
        print("[%s] %s | %s Year %d | Avg: %.2fs" % (
            self.roll_no_str, self.name_str, self.branch_str, 
            self.year_int, self.avg_time_flt))

class EventStruct:
    def __init__(self, ev_str, d_str, v_str):
        self.event_name_str = ev_str
        self.date_str = d_str
        self.venue_str = v_str
        self.participant_arr = [] 
        self.participant_cnt_int = 0

    def register_participant(self, p_member_ptr):
        self.participant_arr.append(p_member_ptr)
        self.participant_cnt_int += 1
        print("Registered %s for %s." % (p_member_ptr.name_str, self.event_name_str))

    def remove_participant_by_roll(self, target_roll_str):
        idx = 0
        while idx < self.participant_cnt_int:
            if self.participant_arr[idx].roll_no_str == target_roll_str:
                print("Removed %s from %s." % (self.participant_arr[idx].name_str, self.event_name_str))
                self.participant_arr.pop(idx)
                self.participant_cnt_int -= 1
                return 1
            idx += 1
        
        print("Error: Participant not found.")
        return 0

    def print_participants(self):
        print("\nParticipants for %s:" % (self.event_name_str))
        if self.participant_cnt_int == 0:
            print("No members registered.")
            return
            
        idx = 0
        while idx < self.participant_cnt_int:
            curr_ptr = self.participant_arr[idx]
            print("- %s (%s)" % (curr_ptr.name_str, curr_ptr.roll_no_str))
            idx += 1

class ClubManager:
    def __init__(self):
        self.member_arr = []
        self.event_arr = []
        self.total_members_int = 0
        self.total_events_int = 0

    def insert_member(self, n_str, r_str, b_str, y_int, c_flt):
        new_node_ptr = MemberStruct(n_str, r_str, b_str, y_int, c_flt)
        self.member_arr.append(new_node_ptr)
        self.total_members_int += 1
        print("Inserted member: %s" % (n_str))

    def get_member_ptr_by_roll(self, search_roll_str):
        idx = 0
        while idx < self.total_members_int:
            if self.member_arr[idx].roll_no_str == search_roll_str:
                return self.member_arr[idx]
            idx += 1
        return None

    def create_event(self, n_str, d_str, v_str):
        new_ev_ptr = EventStruct(n_str, d_str, v_str)
        self.event_arr.append(new_ev_ptr)
        self.total_events_int += 1
        print("Created event: %s" % (n_str))

    def print_all_members(self):
        print("\nAll Club Members:")
        idx = 0
        while idx < self.total_members_int:
            self.member_arr[idx].print_details()
            idx += 1
            
    def print_all_events(self):
        print("\nAll Club Events:")
        idx = 0
        while idx < self.total_events_int:
            ev_ptr = self.event_arr[idx]
            print("%s | %s @ %s | %d Participants" % (
                ev_ptr.event_name_str, ev_ptr.date_str, ev_ptr.venue_str, ev_ptr.participant_cnt_int))
            idx += 1

    def print_top_three(self):
        print("\nTop 3 Fastest Cubers:")
        if self.total_members_int == 0:
            print("Club is empty.")
            return

        temp_arr = []
        c_idx = 0
        while c_idx < self.total_members_int:
            temp_arr.append(self.member_arr[c_idx])
            c_idx += 1
        
        i = 0
        while i < self.total_members_int:
            j = 0
            while j < self.total_members_int - i - 1:
                if temp_arr[j].avg_time_flt > temp_arr[j + 1].avg_time_flt:
                    tmp_ptr = temp_arr[j]
                    temp_arr[j] = temp_arr[j + 1]
                    temp_arr[j + 1] = tmp_ptr
                j += 1
            i += 1

        limit_int = 3
        if self.total_members_int < 3:
            limit_int = self.total_members_int
            
        k = 0
        while k < limit_int:
            print("%d. %s - %.2fs" % (k + 1, temp_arr[k].name_str, temp_arr[k].avg_time_flt))
            k += 1

    def print_avg_time(self):
        if self.total_members_int == 0:
            return
        total_flt = 0.0
        idx = 0
        while idx < self.total_members_int:
            total_flt += self.member_arr[idx].avg_time_flt
            idx += 1
        
        avg_flt = total_flt / self.total_members_int
        print("\nOverall Club Average: %.2fs" % (avg_flt))

    def print_branch_counts(self):
        print("\nBranch-Wise Count:")
        unique_branches = []
        counts = []
        unique_cnt = 0
        
        idx = 0
        while idx < self.total_members_int:
            curr_branch = self.member_arr[idx].branch_str
            found = 0
            b_idx = 0
            while b_idx < unique_cnt:
                if unique_branches[b_idx] == curr_branch:
                    counts[b_idx] += 1
                    found = 1
                    break
                b_idx += 1
                
            if found == 0:
                unique_branches.append(curr_branch)
                counts.append(1)
                unique_cnt += 1
            idx += 1
            
        p_idx = 0
        while p_idx < unique_cnt:
            print("%s: %d member(s)" % (unique_branches[p_idx], counts[p_idx]))
            p_idx += 1

    #bonus features
    def search_by_branch_name(self, target_branch_str):
        print("\nMembers in %s:" % (target_branch_str))
        found_int = 0
        idx = 0
        while idx < self.total_members_int:
            if self.member_arr[idx].branch_str == target_branch_str:
                self.member_arr[idx].print_details()
                found_int = 1
            idx += 1
            
        if found_int == 0:
            print("No members found in this branch.")

    def export_data(self, file_name_str):
        file_ptr = open(file_name_str, "w")
        file_ptr.write("SPEEDCUBING CLUB MEMBERS\n\n")
        
        idx = 0
        while idx < self.total_members_int:
            curr_ptr = self.member_arr[idx]
            out_str = "Roll: %s | Name: %s | Branch: %s | Year: %d | Avg: %.2fs\n" % (
                curr_ptr.roll_no_str, curr_ptr.name_str, curr_ptr.branch_str, 
                curr_ptr.year_int, curr_ptr.avg_time_flt)
            file_ptr.write(out_str)
            idx += 1
        
        file_ptr.close()
        print("\nExported member data to %s" % (file_name_str))


if __name__ == "__main__":
    system_ptr = ClubManager()

    while 1:
        print("\nMENU")
        print("1. Add Member")
        print("2. Create Event")
        print("3. Register Member for Event")
        print("4. Display All Members")
        print("5. Display All Events")
        print("6. Display Event Participants")
        print("7. Display Top 3 Cubers")
        print("8. Display Average Club Time")
        print("9. Display Branch Counts")
        print("10. Search Member by Branch")
        print("11. Export Data to File")
        print("0. Exit")
        
        choice_str = input("Enter choice: ")

        if choice_str == "1":
            n_str = input("Name: ")
            r_str = input("Roll Number: ")
            b_str = input("Branch: ")
            # Typecasting raw inputs to match C types
            y_int = int(input("Year: "))
            c_flt = float(input("Average Time: "))
            system_ptr.insert_member(n_str, r_str, b_str, y_int, c_flt)

        elif choice_str == "2":
            n_str = input("Event Name: ")
            d_str = input("Date: ")
            v_str = input("Venue: ")
            system_ptr.create_event(n_str, d_str, v_str)

        elif choice_str == "3":
            if system_ptr.total_events_int == 0:
                print("No events exist. Create one first.")
                continue
                
            r_str = input("Enter Member Roll Number: ")
            member_ptr = system_ptr.get_member_ptr_by_roll(r_str)
            
            if member_ptr == None:
                print("Member not found.")
            else:
                # Assuming registration to the first event for simplicity, 
                # or you could prompt for an event index
                system_ptr.event_arr[0].register_participant(member_ptr)

        elif choice_str == "4":
            system_ptr.print_all_members()

        elif choice_str == "5":
            system_ptr.print_all_events()

        elif choice_str == "6":
            if system_ptr.total_events_int == 0:
                print("No events exist.")
            else:
                system_ptr.event_arr[0].print_participants()

        elif choice_str == "7":
            system_ptr.print_top_three()

        elif choice_str == "8":
            system_ptr.print_avg_time()

        elif choice_str == "9":
            system_ptr.print_branch_counts()

        elif choice_str == "10":
            b_str = input("Enter Branch Name: ")
            system_ptr.search_by_branch_name(b_str)

        elif choice_str == "11":
            system_ptr.export_data("club_members.txt")

        elif choice_str == "0":
            print("Exiting program...")
            break
            
        else:
            print("Invalid choice. Try again.")
````
